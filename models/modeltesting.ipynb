```python
import cv2
from ultralytics import YOLO

model = YOLO("last.pt")
results = model("worker.jpeg")

# Load the original image with OpenCV to draw on it
img = cv2.imread("worker.jpeg")

# Define our custom safety colors in BGR format (OpenCV uses Blue, Green, Red)
GREEN = (0, 255, 0)   # Safe (e.g., Hardhat, Safety Vest)
RED = (0, 0, 255)     # Danger (e.g., NO-Hardhat, Fall-Detected)
BLUE = (255, 0, 0)    # Neutral (e.g., Person)

for result in results:
    for box in result.boxes:
        # Get coordinates
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        
        # Get class ID and name
        class_id = int(box.cls[0].item())
        class_name = model.names[class_id]
        
        # Pick a color based on the class name prefix
        if "NO-" in class_name or "Fall" in class_name:
            box_color = RED
        elif class_name in ["Person", "Ladder"]:
            box_color = BLUE
        else:
            box_color = GREEN # Everything else is safety gear!

        # 1. Draw the bounding box rectangle
        cv2.rectangle(img, (x1, y1), (x2, y2), box_color, thickness=2)
        
        # 2. Add text label above the box
        label = f"{class_name} {box.conf[0].item():.2f}"
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, 2)

# Save your freshly painted image
cv2.imwrite("detected_output.jpg", img)
```
